```python
financials.info()
customers.info()
```

```python
financials['date'] = pd.to_datetime(financials['date'])
customers['transaction_date'] = pd.to_datetime(customers['transaction_date'])
```

```python
financials['profit_margin'] = financials['profit'] / financials['revenue']

financials['month'] = financials['date'].dt.to_period('M').astype(str)
customers['month'] = customers['transaction_date'].dt.to_period('M').astype(str)
```

```python
financials.isnull().sum(), customers.isnull().sum()
```

```python
financials.to_csv('../data/processed/clean_financials.csv', index=False)
customers.to_csv('../data/processed/clean_customers.csv', index=False)
```
